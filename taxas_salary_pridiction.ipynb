# Data Exploration and Visualization

```python
taxas.info()
```

```python
# above data gender is a object so we have to change into category for plotting graph

taxas.loc[:, "GENDER"] = taxas["GENDER"].astype("category")
taxas.loc[:, "ETHNICITY"] = taxas["ETHNICITY"].astype("category")
```

```python
taxas.info()
```

```python
taxas.describe()
```

```python
# gender base monthly income

sns.catplot(data=taxas,x="GENDER",y="MONTHLY",hue="ETHNICITY",kind="bar",errorbar=None)
plt.show()
```

```python
# gender base annual income

sns.catplot(data=taxas,x="GENDER",y="ANNUAL",hue="ETHNICITY",kind="bar",errorbar=None)
plt.show()
```

```python
# hourly and monthly comparision

sns.catplot(data=taxas, x='HRLY RATE', y='MONTHLY', row='ETHNICITY', kind='bar', col='GENDER')
plt.show()
```

```python
#plot for monthly box plot

sns.catplot(data=taxas, x="ETHNICITY", y="MONTHLY", hue="GENDER", kind="box", errorbar=None, height=7, aspect=2)
```

```python
#plot for ANNUAL violin

sns.catplot(data=taxas, x="ETHNICITY", y="ANNUAL", hue="GENDER", kind="violin", errorbar=None, height=7, aspect=2)
```

# Anaylis with groupby

```python
salary=taxas.groupby("AGENCY NAME")
salary.size()
```

```python
salary[["HRLY RATE","HRS PER WK","MONTHLY","ANNUAL"]].agg(["min","max",pd.Series.mean])
```

```python
# max salary in all agencies

salary["ANNUAL"].sum().sort_values(ascending=False).head(3)
```

```python
# minimum salary in all agencies

salary["ANNUAL"].sum().sort_values(ascending=True).head(3)
```

```python
# new groub by

taxas.groupby("ETHNICITY")["ANNUAL"].sum().plot(kind="bar")
```

```python
# highest paid employee top 10

highest_paid = taxas.groupby('FIRST NAME')['MONTHLY'].sum().sort_values(ascending=False).head(10)

print(highest_paid)
```

```python
# lowest paid employee top 10

low_paid = taxas.groupby('FIRST NAME')['MONTHLY'].sum().sort_values(ascending=False).tail(10)

print(low_paid)
```
